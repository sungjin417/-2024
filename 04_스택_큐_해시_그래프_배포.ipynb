# 스택(Stacks)

## 스택의 개념과 원리

### 스택이란 ?

- 자료를 보관할 수 있는 선형구조, 단, 넣을 때에는 한 쪽 끝에서 밀어 넣어야 하고 꺼낼때에는 같은 쪽에서 뽑아 꺼내야 하는 제약이 있음
- 접시가 쌓여 있으면 맨 위에 있는 접시를 들어내 사용하고 새 접시는 맨 위에 쌓는 것처럼 데이터를 이런 방식으로 쌓고 빼는 자료구조가 스택이다.
- push(), pop()과 같은 2가지 주요 연산을 지원하는 요소의 컬렉션으로 사용되는 추상 자료형이다.
  - push(): 요소를 컬렉션에 추가한다.
  - pop(): 아직 제거되지 않은 가장 최근에 삽입된 요소를 제거한다.
- 가장 뒤에 들어온 것이 가장 먼저 나가는 구조라는 뜻으로 LIFO(Last-In-First-Out, 후입선출)로 처리
- 파이썬은 스택 자료형을 별로도 제공하지는 않지만, 리스트가 사실상 스택의 모든 연산 지원

### 스택의 동작

- 스택은 맨 위의 원소만 접근 가능한 것이 특징이다.
- 스택의 맨 위 원소를 스택 탑 또는 스택 탑 원소라 한다.
- 새 원소를 삽입(push)할 때는 스택 탑 바로 윗자리에 원소를 저장한 후 새 원소가 새 스택 탑(top)이 되게 한다.
- 원소를 삭제(pop)할 때는 무조건 탑에 있는 원소를 삭제한 후 바로 아래 원소가 새 스택 탑(top)이 되게 한다.

- 스택 비우기: 입력 문자열을 모두 읽고 나면, 스택에 남아 있는 모든 연산자를 팝하여 결과 리스트에 추가

    (연산자를 만났을 때, 여는 괄호 너머까지 pop 하지 않도록 여는 괄호의 우선순위는 가장 낮게 설정)

- 괄호의 처리

    (예)  ** (A+B)*C -> AB+C* **

# 큐(Queues)


## 큐(Queues) 란?

- 자료를 보관할 수 있는 선형구조, 단 넣을 때에는 한 쪽 끝에서 밀어 넣어야 하고 꺼낼때에는 반대 쪽에서 뽑아 꺼내야 하는 제약이 있음
- 마트에서 계산하기 위해 줄을 서거나 은행등에서 번호표를 뽑고 기다리는 것과 같이 먼저 집어 넣은 데이터가 먼저 나오는 자료구조

-  **FIFO(First In First Out, 선입선출) 처리되는데 스택에 비해 상대적으로 쓰임새가 적지만, 테크(deque) 나 우선순위 큐(Priority Queue)같은 변형들은 여러 분야에서 매우 유용**

- enqueue(), dequeue()과 같은 2가지 주요 연산을 지원하는 요소의 컬렉션으로 사용되는 추상 자료형이다.
  - enqueue(): 요소를 컬렉션에 맨 뒤에 추가한다.
  - dequeue(): 컬렉션에서 가장 앞에 있는 요소를 삭제한다.

- **이외에도 너비우선탐색(Breath-First-Search, BFS)이나 캐시 등을 구현할 때도 널리 사용**

## 환형 큐(Circular Queue)

- 원형 큐(환영 큐)는 FIFO 구조를 지닌다는 점에서 기존 큐와 동일하나 마지막 위치가 시작위치와 연결되어 원형구조를 띤다.

- 기존 큐는 공간이 꽉 차면  더이상 요소를 추가할 수 없었다. 심지어 앞쪽에 요소들이 dequeue() 로 모두 빠져서 공간이 충분히 남는다 해도 추가할 수 방법이 없었다 이를 개선해서 앞쪽에 공간이 남아 있다면 앞쪽에 추가할 수 있도록 재활용 가능한 구조가 바로 원형 큐의 삽입과 삭제 원리 이다

# DFS & BFS 란?

## 그래프(graphs) 란?

- 그래프(Graph)는 연결되어있는 원소간의 관계를 표현한 비선형 자료구조
- 그래프는 연결할 객체를 나타내는 정점(vertex, node)과 객체를 연결하는 간선(edge, link)의 집합으로 구성



image.png

- 그래프의 종류

image.png

#### 그래프의 순회

- 그래프의 순회란 한 정점에서 시작하여 도달 가능한 모든 정점을 방문하는 것
- 대표적인 알고리즘으로 너비우선탐색(BFS Breadth-First Search), 깊이 우선 탐색(Depth-First Search) 있다.
  

##### 깊이 우선 탐색(DFS; Depth-First Search)

- 시작 정점에서 출발하여 한 방향으로 갈 수 있는 만큼 최대한 깊게 탐색한 후, 더 이상 갈 수 없게 되면 마지막으로 만났던 갈림길로 돌아와서 다른 방향으로 탐색을 계속하는 방식

- 주로 스택 또는 재귀 함수 사용

image.png

##### 너비 우선 탐색(BFS; Breadth-First Search)

- 시작 정점에서 가까운 정점부터 먼저 탐색을 진행하는 방식 즉, 시작 정점에서 같은 레벨에 있는 정점들을 모두 방문한 후, 그 다음 레벨로 넘어가는 식으로 탐색을 진행
- 큐 사용

image.png

- 위 방법이 유일한 경우는 아님

- DFS (Depth-First Search)
    - 모든 경로 탐색: 그래프나 트리의 모든 가능한 경로를 탐색할 때 유용합니다.
    - 백트래킹: 퍼즐 해결 등에서 가능한 모든 조합을 시도하면서 해결책을 찾는 경우에 사용됩니다.
- BFS (Breadth-First Search)
    - 최단 경로 탐색: 가중치가 없는 그래프에서 최단 경로를 찾을 때 가장 적합합니다.
